**RDBMS Strenghts**  
- Transactions, OLTP  
- Enforcing Business Rules  
- Structure  
- Good Choices - Postgres, SQLite  
- Strong with Small or Minimum Data  
- Simple Security Controls  
- Fast(at reasonable scale)  
- Lots of tools and solutions  


**Limitations of RDBMS**
- Schema on write  
- High cost of storage  
- Weak support for unstructured data  
- Difficulty of distributed transactions  

**Semi Structured Data**
- Records with labels but no fixed schema  
    a. JSON, XML  
    b. CSV files with headers  
- Records with some structure but no internal labels or schema  
    a. Log Files  
    
- Extracting Content from Semi-Structured Data  
    a. JSON Functions(lik get_json_object)  
    b. XPath functions, for XML strings  
    c. Regular expressions and other string functions, for patterns like log records  
    
**Schema on Read**: A technique in which structure is not enforced on data when it is stored, but is applied to data when it is read.
